```python
import pandas as pd
```

```python
def distancia_igual(df, atributo, qtd_intervalos=-1, should_print=True):
    # Distancia entre cada um:
    # [min + W] [min + 2W] ... [min + nW]
    # n = total de distancias
    # W = (max-min)/n
    df_sorted = df.sort_values(atributo)  # Ordenar a coluna
    quantidade = len(df_sorted[atributo].unique())

    min = df_sorted[atributo].min()
    max = df_sorted[atributo].max()

    n = int(quantidade / 10) if qtd_intervalos < 1 else qtd_intervalos
    w = (max - min) / n

    # Nomes para substituir os numeros
    nomes = []
    for i in range(1, n + 1):
        nomes.append((atributo + "_range_" + str(i)))

    # Algoritmo
    intervalos_n = []
    for i in range(0, n + 1):
        intervalos_n = intervalos_n + [min + w * i]

    intervalos = []
    for i in range(0, n):
        temp = []
        for elemento in df_sorted[atributo]:
            if intervalos_n[i] <= elemento < intervalos_n[i + 1] or elemento == max:
                temp += [elemento]
                df[atributo] = df[atributo].replace(elemento, nomes[i])
        intervalos += [temp]

    # Prints
    if should_print:
        print("\nMin: ", min)
        print("\nMax: ", max)
        print("\nqtd total: ", quantidade)
        print("\nNumero de intervalos: ", n)
        print("\nValor de W: ", w)

        print("\nIntervalos:\n")
        for i in intervalos:
            print(i)

        print("\nNomes: " + str(nomes) + "\n")
        print(df[atributo])

    return df[atributo]


if __name__ == '__main__':
    df = pd.read_csv("/content/balanced.csv", encoding="UTF-8", delimiter=",")
    df = df.round(1)
    df = df.drop(columns=df.columns[0])

    df["precipitation"] = distancia_igual(df, "precipitation", 6)
    df["temp_max"] = distancia_igual(df, "temp_max", 5)
    df["temp_min"] = distancia_igual(df, "temp_min", 5)
    df["wind"] = distancia_igual(df, "wind", 4)

    df.to_csv("binned.csv")
```
